```python
import docx
from docx import Document
doc = docx.Document("Tata_Safari_2021_Car_Review.docx")
```

```python
paras = [p.text for p in doc.paragraphs if p.text]   
print(paras)
```

We have got the paragraphs from the doc file as above. Now just extract the sentences from all the paragraphs using nltk

```python
from nltk import tokenize
sentences = []
for p in paras:
    sentences_in_para = tokenize.sent_tokenize(p)
    for s in sentences_in_para:
        print('----->')
        print(s)
        sentences.append(s)
```

```python
print(sentences)
```

Now we have got the individual sentences from the doc file as above. Next use nltk(vadar_lexicon) to extract the positive sentence.

```python
# If you haven’t already, download the lexicon
import nltk
# from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
# Initialize the VADER sentiment analyzer
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
analyzer.polarity_scores("If you should hit a rough patch on your holiday outings, Tata has you covered there too, with the Rough Road Mode which optimizes the vehicle's behavior on broken roads.")
```

```python
positive_sentences = []
for s in sentences:
    print(s)
    score = analyzer.polarity_scores(s)
    print(score)
    if score['pos'] > 0.2:
        print('>---positive---<')
        positive_sentences.append(s)
```

```python
for p in positive_sentences:
    print("..." + p)
```
